# import necessory packeges

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Load the data into datafarme

```python
data =pd.read_csv("tv ads.csv")
```

#view the first 5 rows in the dataframe

```python
data.head()
```

# to check the nul values

```python
data.isna().sum()
```

# spliting features 

```python
X = data[["sales($)"]]
```

```python
Y= data[["TV ads"]]
```

# to see how the data spread using scatter plot 

```python
plt.scatter(X,Y,color = 'red')
```

# using Regression algorithm

```python
model =LinearRegression()
```

```python
# training the data
```

```python
model.fit(X,Y)
```

```python
# check the accuracy score
```

```python
model.score(X,Y)
```

```python
# predict the outcomes
```

```python
pred_y = model.predict(X)
```

```python
pred_y
```

```python
plt.scatter(X,Y)
plt.plot(X,pred_y)
```
